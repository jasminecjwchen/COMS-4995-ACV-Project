Restructure the JSON file train_json, so that it fits our file structure and maps to a 5-class classifier

```python
import json
import os
```

```python
original_train_json = '/Users/jasminecjwchen/Documents/GitHub/COMS-4995-ACV-Project/preprocessing/original_test_label.json'
```

```python
# Load the original JSON labels
with open(original_train_json, 'r') as file:
    original_labels = json.load(file)
```

```python
# Define the mapping from original labels to the new classification scheme
label_mapping = {0: 0, 1: 1, 2: 1, 3: 1, 4: 2, 5: 2, 6: 2, 7: 3, 8: 3, 9: 3, 10: 4}
```

```python
# New base directory where images are now stored
new_base_dir = '/Users/jasminecjwchen/Documents/GitHub/COMS-4995-ACV-Project/preprocessing'
```

```python
updated_labels = {}
for img_path, label_array in original_labels.items():
    original_label = label_array[:41]  # the label is at position 40
    
    # update the final label to the mappin above
    original_label[40] = label_mapping[original_label[40]]
    
    filename = os.path.basename(img_path)
    updated_labels[filename] = original_label
```

```python
# Save the updated labels to a new JSON file
with open('/Users/jasminecjwchen/Documents/GitHub/COMS-4995-ACV-Project/preprocessing/test_labels v2.json', 'w') as outfile:
    json.dump(updated_labels, outfile)
```
